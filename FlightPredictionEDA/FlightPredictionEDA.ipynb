```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("flight_price.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df[["Date", "Month", "Year"]] = df["Date_of_Journey"].str.split("/", expand=True)
df.head()
```

```python
df["Date"] = df["Date"].astype(int)
df["Month"] = df["Month"].astype(int)
df["Year"] = df["Year"].astype(int)
df.dtypes
```

```python
df.head()
```

```python
df["Arrival_Hour"] = df["Arrival_Time"].str.split(":").str[0].astype(int)
df["Arrival_Minute"] = (
    df["Arrival_Time"].str.split(":").str[1].str.split(" ").str[0].astype(int)
)
```

```python
df.head()
```

```python
df.drop({"Date_of_Journey", "Arrival_Time"}, axis=1, inplace=True)
df.head()
```

```python
df["Total_Stops"] = df["Total_Stops"].map(
    {"non-stop": 0, "2 stops": 2, "1 stop": 1, "3 stops": 3, "4 stops": 4}
)
```
