```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Univariate Analysis

# Histogram
df.hist(figsize=(10, 10), bins=50)
plt.tight_layout()
plt.show()
```

```python
df.columns = df.columns.str.strip()
df.columns = df.columns.str.upper()
```

```python
df.columns
```

```python
# Save the cleaned dataset
df.to_csv("algerian_forest_fires_cleaned.csv", index=False)
```

```python
df.info()
```

```python
df.hist(figsize=(10, 10), bins=50)
plt.tight_layout()
plt.show()
```

```python
# Bivariate Analysis

# Scatter plot
sns.pairplot(df, hue="CLASSES")
plt.show()
```

```python
# heatmap
sns.heatmap(
    df.drop({"CLASSES"}, axis=1).corr(),
    annot=True,
    annot_kws={"fontsize": 8},
    cmap="coolwarm",
    fmt=".2f",
)
plt.figure(figsize=(50, 50), dpi=500)
plt.show()
```

```python
print(df.drop({"CLASSES"}, axis=1).corr())
```

```python
# Multivariate Analysis

# Box plot for all features
plt.figure(figsize=(15, 10))
df.boxplot(
    column=["TEMPERATURE", "RH", "WS", "RAIN", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]
)
plt.show()
```

```python
# Montly distribution of fires
sns.countplot(x="MONTH", hue="CLASSES", data=df)
plt.show()
```
